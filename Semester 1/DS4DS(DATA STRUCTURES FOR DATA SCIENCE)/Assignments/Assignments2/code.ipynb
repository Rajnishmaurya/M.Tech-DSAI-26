```python
class Node:
    def __init__(self,element,prev=None,next=None):
        self.element=element
        self.prev=prev
        self.next=next
```

```python
class Positional_List:
    def __init__(self):
        self.head=None
        self.tail=None
        self.size=0
    
    def length(self):
        return self.size
    
    def is_empty(self):
        return self.size==0
    
    def first(self):
        if not self.is_empty:
            return None
        return self.head
    
    def last(self):
        if self.is_empty():
            return None
        return self.tail
    
    def before(self,p):
        if p==self.head:
            return None
        return p.prev
    
    def after(self,p):
        if p==self.tail:
            return None
        return p.next
    
    def add_first(self,key):
        node=Node(key,None,self.head)
        if self.is_empty():
            self.tail=self.head=node
        else:
            self.head.prev=node
            self.head=node
        self.size+=1
        return node
    
    def add_last(self,key):
        node=Node(key,self.tail,None)   
        if self.is_empty():
            self.head=self.tail=node
        else:
            self.tail.next=node
            self.tail=node
        self.size+=1
        return node
    
    def add_before(self,p,e):
        if self.head is None:
            return self.add_first(e)
        
        node=Node(e,p.prev,p)
        p.prev.next=node
        p.prev=node
        self.size+=1
        return node
    
    def add_after(self,a,e):
        if self.tail is None:
            return self.add_last(e)
        node=Node(e,a,a.next)
        a.next.prev=node
        a.next==node
        self.size+=1
        return node
    
    def replace(self,p,e):
        old_element=p.element
        p.element=e
        return old_element
    
    def delete(self,p):
        if p==self.head:
            self.head=self.head.next
            if self.head:
                self.head.prev=None
        elif p==self.tail:
            self.tail=p.prev
            if self.tail:
                self.tail.next=None
        else:
            p.prev.next=p.next
            p.next.prev=p.prev
        self.size-=1
        return p.element
    
    def iter(self):
        temp=self.head
        
        while temp:
            yield temp.element
            temp=temp.next

    def element(self,p):
        return p.element
    
    def __str__(self):
        return str(list(self.iter()))                            
        
                     
            
        
    
     
```
