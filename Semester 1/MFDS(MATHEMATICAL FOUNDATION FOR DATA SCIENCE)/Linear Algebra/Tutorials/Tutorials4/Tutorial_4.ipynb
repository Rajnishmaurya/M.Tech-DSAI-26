# Eigen Values and Vectors

### Definition
- **Eigenvalue**: For a square matrix \( A \), an eigenvalue \( $\lambda$ \) is a scalar such that there exists a non-zero vector \( $\mathbf{v}$ \) (called the eigenvector) which satisfies the equation:
  $$
  A \mathbf{v} = \lambda \mathbf{v}
  $$
- **Eigenvector**: A non-zero vector \( $\mathbf{v}$ \) associated with an eigenvalue \( $\lambda$ \) of a matrix \( A \), which is scaled by \( $\lambda$ \) when the matrix \( A \) is applied to it.

### Example
Consider a matrix \( A \):
$$
A = \begin{bmatrix}
4 & 1 \\
2 & 3
\end{bmatrix}
$$

To find the eigenvalues \( $\lambda$ \) and eigenvectors \( $\mathbf{v}$ \), solve the characteristic equation:
$$
\text{det}(A - \lambda I) = 0
$$
where \( I \) is the identity matrix. This gives us:
$$
\text{det}\begin{bmatrix}
4 - \lambda & 1 \\
2 & 3 - \lambda
\end{bmatrix} = 0
$$

### Solution
Solving the characteristic polynomial will provide eigenvalues. Substitute each eigenvalue into $(A - \lambda I) \mathbf{v} = 0$ to find the corresponding eigenvectors.

### Step 1: Find the Eigenvalues
The determinant is:
$$
\text{det}\begin{bmatrix}
4 - \lambda & 1 \\
2 & 3 - \lambda
\end{bmatrix} = (4 - \lambda)(3 - \lambda) - 2 \times 1 = \lambda^2 - 7\lambda + 10 = 0
$$

### Step 2: Solve the Characteristic Polynomial
Solve the quadratic equation:
$$
\lambda^2 - 7\lambda + 10 = 0
$$

The roots of this equation, which are the eigenvalues, are given by:
$$
\lambda = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$
where \( a = 1 \), \( b = -7 \), and \( c = 10 \). Substituting these values:
$$
\lambda = \frac{7 \pm \sqrt{49 - 40}}{2}
$$
$$
\lambda = \frac{7 \pm \sqrt{9}}{2}
$$
$$
\lambda = \frac{7 \pm 3}{2}
$$

Thus, the eigenvalues are:
$$
\lambda_1 = \frac{7 + 3}{2} = 5
$$
$$
\lambda_2 = \frac{7 - 3}{2} = 2
$$

### Step 3: Find the Eigenvectors

#### Eigenvector for \( $\lambda_1$ = 5 \):
Substitute \( $\lambda$ = 5 \) into \( A - $\lambda$ I \):
$$
A - 5I = \begin{bmatrix}
4 - 5 & 1 \\
2 & 3 - 5
\end{bmatrix} = \begin{bmatrix}
-1 & 1 \\
2 & -2
\end{bmatrix}
$$

Solve the equation \( (A - 5I) $\mathbf{v}$ = 0 \):
$$
\begin{bmatrix}
-1 & 1 \\
2 & -2
\end{bmatrix}
\begin{bmatrix}
v_1 \\
v_2
\end{bmatrix} = \begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

This simplifies to:
$$
- v_1 + v_2 = 0
$$
$$
2v_1 - 2v_2 = 0
$$

Both equations simplify to \( $v_1$ = $v_2$ \). So, the eigenvector corresponding to \( $\lambda_1$ = 5 \) is:
$$
\mathbf{v}_1 = \begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

#### Eigenvector for \( $\lambda_2 = 2$ \):
Substitute \( $\lambda$ = 2 \) into \( A - $\lambda$ I \):
$$
A - 2I = \begin{bmatrix}
4 - 2 & 1 \\
2 & 3 - 2
\end{bmatrix} = \begin{bmatrix}
2 & 1 \\
2 & 1
\end{bmatrix}
$$

Solve the equation \( (A - 2I) $\mathbf{v}$ = 0 \):
$$
\begin{bmatrix}
2 & 1 \\
2 & 1
\end{bmatrix}
\begin{bmatrix}
v_1 \\
v_2
\end{bmatrix} = \begin{bmatrix}
0 \\
0
\end{bmatrix}
$$

This simplifies to:
$$
2v_1 + v_2 = 0
$$

These equations simplify to \( $v_2$ = $-2v_1$ \). So, the eigenvector corresponding to \( $\lambda_2$ = 2 \) is:
$$
\mathbf{v}_2 = \begin{bmatrix}
1 \\
-2
\end{bmatrix}
$$

### Summary
For the matrix  A = $\begin{bmatrix} 4 & 1 \\ 2 & 3 \end{bmatrix}$ , the eigenvalues and corresponding eigenvectors are:
-  $\lambda_1$ = 5  with eigenvector  $\mathbf{v}_1$ = $\begin{bmatrix} 1 \\ 1 \end{bmatrix}$ 
-  $\lambda_2$ = 2  with eigenvector  $\mathbf{v}_2$ = $\begin{bmatrix} 1 \\ -2 \end{bmatrix}$ 




```python
import numpy as np
A = np.array([[4, 1],
             [2, 3]])

eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors (as columns):")
print(eigenvectors)


```

## Singular Value Decomposition (SVD)

### Definition
- **Singular Value Decomposition (SVD)**: For any \( m $\times$ n \) matrix \( A \), the SVD is a factorization of \( A \) into three matrices:
  $$
  A = U \Sigma V^T
  $$
  where:
  -  U  is an \( $m \times m$ \) orthogonal matrix (left singular vectors),
  -  $\Sigma$  is an \( $m \times n$ \) diagonal matrix with non-negative real numbers (singular values) on the diagonal,
  -  $V^T$ is an \( $n \times n$ \) orthogonal matrix (right singular vectors).

### Example
Given a matrix \( A \):
$$
A = \begin{bmatrix}
1 & 2 & 3\\
4 & 5 & 6 
\end{bmatrix}
$$

The SVD of \( A \) involves finding U , $\Sigma$, and  $V^T$ such that $A = U \Sigma V^T$.

### Properties
- The singular values in $\Sigma$ are the square roots of the eigenvalues of  $A^T A$ or $A A^T$.
- SVD is useful for solving linear least squares problems, dimensionality reduction, and data compression.


```python
A = np.array([[1, 2, 3],
              [4, 5, 6]])


U, Sigma, VT = np.linalg.svd(A, full_matrices=True)
Sigma_matrix = np.zeros(A.shape)
k = min(A.shape)
Sigma_matrix[:k, :k] = np.diag(Sigma)

print("U:")
print(U)

print("\nSigma:")
print(Sigma)

print('\nSigma as a diagonal matrix:')
print(Sigma_matrix)

print("\nVT:")
print(VT)
```

```python
U, Sigma, VT = np.linalg.svd(A, full_matrices=False)

print("U:")
print(U)

print("\nSigma:")
print(Sigma)

print('\nSigma as a diagonal matrix:')
Sigma_matrix = np.diag(Sigma)
print(Sigma_matrix)

print("\nVT:")
print(VT)
```

```python
print(VT@VT.T)
print(VT.T@VT)
```

## Eigen Decomposition

- For a square matrix \( A \) that is diagonalizable, there exists a matrix \( P \) of eigenvectors and a diagonal matrix \( D \) of eigenvalues such that:
  $$
  A = P D P^{-1}
  $$


```python
# Eigen Decomposition
A = np.array([[1, 2],
              [2, 1]])

eigenvalues, eigenvectors = np.linalg.eig(A)
print("Eigenvalues:")
print(eigenvalues)

print("\nEigenvectors (as columns):")
print(eigenvectors)

print("\nReconstructing A:")
A_reconstructed = eigenvectors @ np.diag(eigenvalues) @ np.linalg.inv(eigenvectors)
print(A_reconstructed)
```

# Pseudo-Inverse

## Definition
The **pseudo-inverse** (or Moore-Penrose inverse) of a matrix A is a generalization of the matrix inverse that can be applied to any matrix, not just square or invertible matrices. It is denoted as $A^+$ and satisfies the following conditions for a matrix \( A \):

1. 
$$
A A^+ A = A
$$

2. 
$$
A^+ A A^+ = A^+
$$

3. 
$$
(A A^+)^T = A A^+
$$

4. 
$$
(A^+ A)^T = A^+ A
$$

## Calculation

### For a General Matrix
If A is an \( $m \times n$ \) matrix, the pseudo-inverse $A^+$ can be computed using the Singular Value Decomposition (SVD) of  A :

1. **Compute SVD**:
   $$
   A = U \Sigma V^T
   $$
   where:
   - U is an \( $m \times m$ \) orthogonal matrix (left singular vectors),
   - $\Sigma$ is an \( $m \times n$ \) diagonal matrix with singular values on the diagonal,
   - $V^T$ is an \( $n \times n$ \) orthogonal matrix (right singular vectors).

2. **Compute the Pseudo-Inverse**:
   - Invert the non-zero singular values in \( $\Sigma$ \) to get \( $\Sigma^+$ \).
   - The pseudo-inverse \( $A^+$ \) is given by:
     $$
     A^+ = V \Sigma^+ U^T
     $$
   where \( $\Sigma^+$ \) is the matrix obtained by taking the reciprocal of each non-zero singular value in \( $\Sigma$ \) and transposing the resulting matrix.

## Properties
- The pseudo-inverse is unique and always exists for any matrix.
- For a matrix  A that is square and invertible, the pseudo-inverse is the same as the regular inverse.
- **Least-Squares Solution**: For a matrix  A that is over-determined (rows > columns), the pseudo-inverse provides a least-squares solution to the system $A \mathbf{x} = \mathbf{b} $:
  $$
  \mathbf{x} = A^+ \mathbf{b}
  $$
  This minimizes the error $\|A \mathbf{x} - \mathbf{b}\|_2$.

- **Least-Norm Solution**: For a matrix A that is not square and has more rows than columns (under-determined system), the pseudo-inverse provides a least-norm solution $\mathbf{x}$ to the system $A \mathbf{x} = \mathbf{b}$:
  $$
  \mathbf{x} = A^+ \mathbf{b}
  $$
  This solution has the minimum possible \( $\| \mathbf{x} \|_2$ \) while satisfying $A \mathbf{x} = \mathbf{b}$.

```python
import numpy as np

# Original matrix
A = np.array([[1, 2, 3],
              [4, 5, 6]])

print('Matrix ATA:')
print(A.T @ A)

## Use (ATA)^-1 AT to calculate the pseudo-inverse of A
# A_pseudo_inverse = np.linalg.inv(A.T @ A) @ A.T         # Does not work for singular matrices
# print('Pseudo-inverse of A:')
# print(A_pseudo_inverse)
```

```python
# Perform Singular Value Decomposition (SVD)
U, Sigma, VT = np.linalg.svd(A, full_matrices=False)

# Create the pseudo-inverse of the Sigma matrix
Sigma_pinv = np.zeros((VT.shape[0], U.shape[0]))  # Create a zero matrix with the appropriate shape
Sigma_pinv[:len(Sigma), :len(Sigma)] = np.diag(1/Sigma)  # Invert the diagonal singular values

# Calculate the pseudo-inverse of A
A_pseudo_inverse = VT.T @ Sigma_pinv @ U.T

# Print the pseudo-inverse of A
print('Pseudo-inverse of A:')
print(A_pseudo_inverse)
```

```python
print('\nA times A pseudo-inverse:')
print(A@A_pseudo_inverse)
```

```python
print('A pseudo-inverse times A:')
print(A_pseudo_inverse@A)
print(np.linalg.det(A_pseudo_inverse@A))
```

# Viewing SVD as Transformations

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a 2x3 matrix (3D to 2D)
A = np.array([[1, 3, 2],
              [2, 1, 2]])


# Perform SVD
U, Sigma, Vt = np.linalg.svd(A)
Sigma_matrix = np.zeros((2, 3))
np.fill_diagonal(Sigma_matrix, Sigma)

# Generate a unit sphere
phi = np.linspace(0, np.pi, 100)
theta = np.linspace(0, 2 * np.pi, 100)
phi, theta = np.meshgrid(phi, theta)

x = np.sin(phi) * np.cos(theta)
y = np.sin(phi) * np.sin(theta)
z = np.cos(phi)
sphere = np.array([x.flatten(), y.flatten(), z.flatten()])

# Apply the 3x2 matrix to project from 3D to 2D
projected = A @ sphere

# Reshape the result to match the grid for plotting
proj_x = projected[0, :].reshape(x.shape)
proj_y = projected[1, :].reshape(y.shape)
proj_z = np.zeros_like(z)  # Since we're projecting into 2D

# Plotting
fig = plt.figure(figsize=(18, 14))

# Original 3D unit sphere
ax1 = fig.add_subplot(331, projection='3d')
ax1.plot_surface(x, y, z, color='r', alpha=0.5)
ax1.set_title('Original 3D Unit Sphere')

# After applying V^T and Sigma (2D projection)
ax2 = fig.add_subplot(333, projection='3d')
ax2.plot_surface(proj_x, proj_y, proj_z, color='g', alpha=0.5)
ax2.set_title('After Projection to 2D')

# Final transformation in 2D space
ax3 = fig.add_subplot(338)
ax3.plot(proj_x.flatten(), proj_y.flatten(), 'g')
ax3.set_title('Resulting 2D Ellipse')
ax3.set_aspect('equal')

# Apply the steps of SVD to the unit sphere
Vt_sphere = Vt.T @ sphere  # V^T step
Sigma_sphere = Sigma_matrix @ Vt_sphere  #  step
U_sphere = U @ Sigma_sphere  # U step

# VT to sphere
ax4 = fig.add_subplot(334, projection='3d')
ax4.plot(Vt_sphere[0, :], Vt_sphere[1, :], Vt_sphere[2,:], 'r', alpha=0.5)
ax4.set_title('After Vt ')
ax4.set_aspect('auto')

# Sigma to VT sphere
ax5 = fig.add_subplot(335, projection='3d')
ax5.plot(Sigma_sphere[0, :], Sigma_sphere[1, :], 0, 'cyan', alpha=0.5)
ax5.set_title('After Sigma ')
ax5.set_aspect('auto')

# U to Sigma sphere
ax6 = fig.add_subplot(336, projection='3d')
ax6.plot(U_sphere[0, :], U_sphere[1, :], 0, 'g', alpha=0.5)
ax6.set_title('After U ')
ax6.set_aspect('auto')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define a 3x2 matrix (2D to 3D)
A = np.array([[2, 1],
              [1, 3],
              [2, 2]])

# Perform SVD
U, Sigma, Vt = np.linalg.svd(A)
Sigma_matrix = np.zeros((3, 2))
Sigma_matrix[:2, :2] = np.diag(Sigma)

# Generate a unit circle in 2D
theta = np.linspace(0, 2 * np.pi, 100)
circle = np.array([np.cos(theta), np.sin(theta)])

# Apply V^T (rotation/reflection in 2D)
Vt_circle = Vt @ circle

# Apply Sigma (scaling)
Sigma_circle = Sigma_matrix @ Vt_circle

# Apply U (final rotation/reflection in 3D)
U_circle = U @ Sigma_circle

# Plotting
fig = plt.figure(figsize=(18, 10))

# Original 2D unit circle
ax1 = fig.add_subplot(231)
ax1.plot(circle[0, :], circle[1, :], 'r')
ax1.set_title('Original 2D Unit Circle')
ax1.set_aspect('equal')
ax1.grid(True)

# after applying V^T
ax2 = fig.add_subplot(234)
ax2.plot(Vt_circle[0, :], Vt_circle[1, :], 'g')
ax2.set_title('After Applying V^T and Sigma')
ax2.set_aspect('equal')
ax2.grid(True)


# ellipsoid after applying Sigma
ax3 = fig.add_subplot(235, projection='3d')
ax3.plot(Sigma_circle[0, :], Sigma_circle[1, :], Sigma_circle[2, :], 'b')
ax3.set_title('Final Ellipsoid after Applying U')
ax3.set_aspect('auto')
ax3.grid(True)

# ellipsoid after applying U
ax4 = fig.add_subplot(236, projection='3d')
ax4.plot(U_circle[0, :], U_circle[1, :], U_circle[2, :], 'b')
ax4.set_title('Ellipsoid after Applying Sigma')
ax4.set_aspect('auto')
ax4.grid(True)

# Directly applying A to the unit circle
A_circle = A @ circle
ax5 = fig.add_subplot(233 , projection='3d')
ax5.plot(A_circle[0, :], A_circle[1, :], A_circle[2,:], 'r')
ax5.set_title('Matrix A applied to the unit circle')
ax5.set_aspect('auto')
ax5.grid(True)


plt.show()
```

# SVD for Data Compression

```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os
plt.rcParams['figure.figsize'] = [16, 8]

# Read the image
A = imread(os.path.join('DATA','dog.jpg'))
X = np.mean(A, -1); # Convert RGB to grayscale

# Space for storing X is proportional to the number of pixels
storage_X = X.size

# Visualize the original image
img = plt.imshow(X)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
# Perform SVD
U, S, VT = np.linalg.svd(X,full_matrices=False)
S = np.diag(S)

# Rank-r approximation
r_values = [5, 20, 100]

# Create subplots
fig, axes = plt.subplots(1, len(r_values) + 1, figsize=(15, 5))

for i, r in enumerate(r_values):
    # Construct approximate image using r largest singular values and corresponding singular vectors
    X_approx = U[:, :r] @ S[:r, :r] @ VT[:r, :]

    storage_X_approx = U[:, :r].size + S[:r, :r].size + VT[:r, :].size
    storage_percent = (storage_X_approx / storage_X) * 100
    
    # Plot the approximate image
    ax = axes[i]
    img = ax.imshow(X_approx, cmap='gray')
    ax.axis('off')
    ax.set_title(f'r = {r}, Storage = {storage_percent:.2f} % ')

# Plot the original image
ax = axes[len(r_values)]
img = ax.imshow(X, cmap='gray')
ax.axis('off')
ax.set_title('Original')


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Plot singular values
axes[0].semilogy(np.diag(S))
axes[0].set_title('Singular Values')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Singular Value')

# Plot cumulative sum of singular values
cumulative_sum = np.cumsum(np.diag(S)) / np.sum(np.diag(S))
axes[1].plot(cumulative_sum)
axes[1].set_title('Singular Values: Cumulative Sum')
axes[1].set_xlabel('Index')
axes[1].set_ylabel('Cumulative Sum')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Create the original image matrix
n = 1000
q = int(n / 4)
X = np.zeros((n, n))
X[(q - 1):(3 * q), (q - 1):(3 * q)] = 1

# Prepare the figure and axes
fig, axs = plt.subplots(1, 3, figsize=(24, 8))

nAngles = 12
angles = np.linspace(0, 45, nAngles)
angles = [0, 1] + list(angles) 
angles = np.array(angles)

nAngles = len(angles)

cmap = 'hsv'
cm_np = plt.get_cmap(cmap)(np.linspace(0, 1, nAngles))

legend_labels = []

# Loop through rotation angles
for i, angle in enumerate(angles):
    # Rotate the image
    Xrot = transform.rotate(X,angle)
    Xrot[np.nonzero(Xrot)] = i

    # Compute SVD of the rotated image
    U, S, VT = np.linalg.svd(Xrot, full_matrices=False)

    # Plot rotated image
    axs[0].imshow(np.ma.masked_where(Xrot == 0, Xrot), cmap=cmap, vmin=0, vmax=nAngles)
    axs[0].axis('off')
    axs[0].set_title(f'Rotated Images')

    # Plot singular values
    axs[1].semilogy(S, '-o', color=tuple(cm_np[i]), label=f'{angle:.1f}')
    axs[1].set_title('SVD Singular Values')
    axs[1].set_xlabel('Index')
    axs[1].set_ylabel('Singular Value')
    axs[1].grid(True)

    # Plot cumulative sum of singular values
    cumulative_sum = np.cumsum(S) / np.sum(S)
    axs[2].plot(cumulative_sum, color=tuple(cm_np[i]), label=f'{angle:.2f}')
    axs[2].set_title('SVD Singular Values: Cumulative Sum')
    axs[2].set_xlabel('Index')
    axs[2].set_ylabel('Cumulative Sum')
    axs[2].grid(True)


axs[1].legend(title='Rotation Angles', bbox_to_anchor=(1.05, 1), loc='upper left')
# axs[2].legend(title='Rotation Angles', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout
plt.tight_layout()
plt.show()
```
